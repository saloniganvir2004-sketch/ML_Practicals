```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/Students_Marks.csv")
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df['Result'] = df['Result'].map({'Pass': 1, 'Fail': 0})
new_student = np.array([4.5, 55])
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
distances = []
for i in range(len(df)):
    student_features = np.array([df.loc[i, 'Hours_Studied'], df.loc[i, 'Previous_Score']])
    distance = euclidean_distance(student_features, new_student)
    distances.append(distance)
df['Distance'] = distances
k = 3
neighbors = df.sort_values(by='Distance').head(k)
prediction = neighbors['Result'].mode()[0]
print("Nearest Neighbors:\n", neighbors[['Hours_Studied', 'Previous_Score', 'Result', 'Distance']], "\n")
if prediction == 1:
    print("Predicted Result: Pass")
else:
    print("Predicted Result: Fail")
```
