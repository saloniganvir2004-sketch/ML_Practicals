```python
import pandas as pd
import numpy as np
df = pd.read_csv("Spam_Dataset.csv")
print("Dataset:\n", df)
print("\nShape of Dataset:", df.shape)
```

```python
df['Message'] = df['Message'].str.lower()
spam_words = ['free', 'win', 'offer', 'money', 'buy']
df['Spam_Word_Count'] = df['Message'].apply(lambda x: sum(word in x for word in spam_words))
print("\nAfter Adding Spam Word Count:\n", df)
```

```python
df['Predicted_Label'] = np.where(df['Spam_Word_Count'] >= 2, 'spam', 'ham')
print("\nPredicted Results:\n", df[['Message', 'Predicted_Label']])
```

```python
accuracy = (df['Label'] == df['Predicted_Label']).mean() * 100
print("\nModel Accuracy:", accuracy, "%")
```
