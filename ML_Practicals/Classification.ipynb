```python
import pandas as pd
import numpy as np
df = pd.read_csv("Student_Classification.csv")
print("Dataset:\n", df)
print("\nShape of Dataset:", df.shape)
```

```python
df['Result_Numeric'] = np.where(df['Result'] == 'Pass', 1, 0)
print("\nAfter Encoding:\n", df)
```

```python
df['Predicted'] = np.where((df['Hours_Studied'] >= 3) & (df['Attendance'] >= 75), 'Pass', 'Fail')
print("\nPredicted Results:\n", df[['Hours_Studied', 'Attendance', 'Predicted']])
```

```python
accuracy = (df['Result'] == df['Predicted']).mean() * 100
print("\nModel Accuracy:", accuracy, "%")
```
