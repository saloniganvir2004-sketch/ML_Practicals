```python
import pandas as pd
import numpy as np
df = pd.read_csv("NxK_Matrix.csv")
print("Original Matrix:\n", df)
print("\nMatrix Shape:", df.shape)
```

```python
transpose = df.T
print("\nTranspose of Matrix:\n", transpose)
```

```python
col_sum = df.sum()
print("\nSum of Each Column:\n", col_sum)
```

```python
row_mean = df.mean(axis=1)
print("\nMean of Each Row:\n", row_mean)
```

```python
added_matrix = df + df
print("\nMatrix Addition (A + A):\n", added_matrix)
```

```python
A = df.to_numpy()
multiplied = np.dot(A, A.T)
print("\nMatrix Multiplication (A × Aᵗ):\n", multiplied)
```
