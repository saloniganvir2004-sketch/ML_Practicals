```python
import pandas as pd
import numpy as np
df = pd.read_csv("Regression_Dataset.csv")
print("Dataset:\n", df)
print("\nShape of Dataset:", df.shape)
```

```python
x = df['Hours_Studied']
y = df['Marks']
m = np.cov(x, y, bias=True)[0][1] / np.var(x)
c = y.mean() - m * x.mean()
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
df['Predicted_Marks'] = m * x + c
print("\nPredicted Values:\n", df)
```

```python
mse = np.mean((df['Marks'] - df['Predicted_Marks']) ** 2)
print("\nMean Squared Error:", mse)
```
