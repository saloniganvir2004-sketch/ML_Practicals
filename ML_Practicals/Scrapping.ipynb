```python
import pandas as pd
import numpy as np
df = pd.read_csv("Scraping_Dataset.csv")
print("Scraped Data:\n", df)
print("\nShape of Dataset:", df.shape)
```

```python
print("\nProduct Names and Prices:\n", df[['Product_Name', 'Price']])
print("\nAverage Price of All Products:", df['Price'].mean())
```

```python
electronics = df[df['Category'] == 'Electronics']
print("\nElectronics Products:\n", electronics)
```

```python
top_product = df.loc[df['Rating'].idxmax()]
print("\nTop Rated Product:\n", top_product)
```
